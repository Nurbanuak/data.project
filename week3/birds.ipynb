```python
# Veri oluÅŸtur
kategorik_veri = ['A', 'B', 'C', 'D']
veri_degerleri = [30, 45, 15, 10]
# Pasta GrafiÄŸi
plt.pie(veri_degerleri, labels=kategorik_veri, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.axis('equal')  # Dairesel yapma
plt.title('Pasta GrafiÄŸi: Kategorik Veri')
plt.show()
```

```python
# SÃ¼tun GrafiÄŸi
plt.bar(kategorik_veri, veri_degerleri, color=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.xlabel('Kategori')
plt.ylabel('DeÄŸer')
plt.title('SÃ¼tun GrafiÄŸi: Kategorik Veri')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
# Zaman Serisi Verisi OluÅŸtur
tarihler = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
veri = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
# Ã‡izgi GrafiÄŸi
plt.plot(tarihler, veri, marker='o', linestyle='-', color='orange')
plt.xlabel('Tarih')
plt.ylabel('DeÄŸer')
plt.title('Ã‡izgi GrafiÄŸi: Zaman Serisi Verisi')
plt.show()
```

```python
# Alan Dolgulu GrafiÄŸi
plt.fill_between(tarihler, veri, color='skyblue', alpha=0.4)
plt.xlabel('Tarih')
plt.ylabel('DeÄŸer')
plt.title('Alan Dolgulu GrafiÄŸi: Zaman Serisi Verisi')
plt.show()
```
