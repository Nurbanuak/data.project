```python
import requests
from bs4 import BeautifulSoup
from html.parser import HTMLParser
import pandas as pd
import numpy as np 
```

```python
url = "https://realpython.github.io/fake-jobs/"
text = requests.get(url)
```

```python
print(text.status_code)
soup = BeautifulSoup(text.content, 'html.parser') 

# Getting the title tag 
print(soup.title) 

# Getting the name of the tag 
print(soup.title.name) 

# Getting the name of parent tag 
print(soup.title.parent.name) 
```

```python
s = soup.find_all('a') 

print(s)
```

```python
soup = BeautifulSoup(text.content, 'html.parser') 
for link in soup.select('a[href*=fake-jobs]'):
    print(link.get('href'))
```

```python
fake_job_links = [a['href'] for a in soup.find_all('a', href=lambda href: href and 'fake-job' in href)]
print(fake_job_links)
```

```python
s = soup.find('div', class_='card-content')
print(s)
```

```python
jobs = []

for job_div in soup.find_all('div', class_='card-content'):
    job = {}

    #  bal
    job['title'] = job_div.find('h2', class_='title is-5').text.strip()

    # irket ad
    job['company'] = job_div.find('h3', class_='subtitle is-6 company').text.strip()

    # Konum
    job['location'] = job_div.find('p', class_='location').text.strip()

    # lan tarihi
    job['date'] = job_div.find('time').text.strip()

    # Bavuru linki
    job['apply_link'] = job_div.find('a', class_='card-footer-item')['href']

    jobs.append(job)

# Oluturulan veri setini ekrana yazdrma
for job in jobs:
    print(job)
```

```python
import csv
with open('products.csv', 'w', newline='') as file:
    # Step 4: Using csv.writer to write the list to the CSV file
    writer = csv.writer(file)
    writer.writerow(jobs) # Use writerow for single list

# Step 5: The file is automatically closed when exiting the 'with' block
```

```python
python_jobs= soup.find_all(
    "h2",string=lambda text: "python" in text.lower()
    
)
```

```python
for job in python_jobs:
    print(job)
```

```python
python_jobs_elements = [
    h2_element.parent.parent.parent for h2_element in python_jobs

]
```

```python
for job_element in python_jobs_elements:
    links= job_element.find_all('a')
    for link in links:
        print(link.text.strip())
```

```python
for job_element in python_jobs_elements:
    links= job_element.find_all('a')
    for link in links:
        if link.text == "Apply":
            print(link)
```

```python
for job_element in python_jobs_elements:
    links= job_element.find_all('a')
    for link in links:
        if link.text == "Apply":
            link_url= link["href"]
            print(f"Apply here : {link_url}\n")
```
